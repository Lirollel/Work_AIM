```python
file = 'C:\\Users\\KlimovaAnnaA\\Documents\\MyFiles\\Projects\\OCP\\salesUnits.xlsx'
Sheet_in_file = 'salesUnits'

from datetime import date
Output_file = "_".join([str(date.today()), 'Mapping.xlsx'])
Sheet_in_output_file = 'mapping'
```

```python
import pandas as pd
import numpy as np

data_export = pd.read_excel(file, sheet_name = Sheet_in_file)
```

```python
data_Sales = data_export
data_Sales.columns
data_Sales[['id', 'sapId', 'ocpSegment', 'holding', 'registryCountry']]
```

```python
data_sapName = data_Sales.loc[~data_Sales.sapName.isna(), ['sapName', 'id']].rename(columns={'sapName': 'CompName'})
data_name = data_Sales.loc[~data_Sales.name.isna(), ['name', 'id']].rename(columns={'name': 'CompName'})
data_name_to_id = pd.concat([data_sapName, data_name], axis = 0).reset_index(drop=True)

# Проверка на наличие пустых строк
assert set(data_name_to_id.CompName.isna().values) == {False}
assert set(data_name_to_id.id.isna().values) == {False}

# Удаление дубликатов CompName
data_name_to_id = data_name_to_id.drop_duplicates(subset='CompName', keep='first')
```

```python
data_name_to_id.loc[data_name_to_id.CompName.str.contains('RUEMO'), 'id'] = 'RUEMO'
data_name_to_id.loc[data_name_to_id.CompName.str.contains('RUESM'), 'id'] = 'RUESM'
```

```python
data_name_to_id.to_excel(Output_file, sheet_name=Sheet_in_output_file, index=False)
```
